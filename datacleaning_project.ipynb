```python
import pandas as pd
```

```python
df = pd.read_csv("Independent_Medical_Reviews.csv")
```

# 1. Filling the missing categorical values with 'Unknown'

```python
df['Diagnosis Category'] = df['Diagnosis Category'].fillna('Unknown')
df['Diagnosis Sub Category'] = df['Diagnosis Sub Category'].fillna('Unknown')
df['Treatment Category'] = df['Treatment Category'].fillna('Unknown')
df['Treatment Sub Category'] = df['Treatment Sub Category'].fillna('Unknown')
df['Age Range'] = df['Age Range'].fillna('Unknown')
df['Patient Gender'] = df['Patient Gender'].fillna('Unknown')
```

# 2. Standardizing the age range (fixing '11_20' to '11-20')

```python
df['Age Range'] = df['Age Range'].str.replace('_', '-')
```


# 3. Filling the missing findings 

```python
df['Findings'] = df['Findings'].fillna('No findings provided')
```

```python
#Removing all whitespaces in string columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

#Checking for duplicates 
df = df.drop_duplicates(subset=['Reference ID'])
```

# Saving cleaned dataset

```python
df.to_csv('cleaned_Independent_Medical_Reviews.csv', index=False)
#To verify
print(df.isnull().sum())
```
