1. Guessing Game

```python
import random
right_number = random.randint(1, 10)
guess_number = int(input("Guess a number from 1-10:"))
if guess_number == right_number:
    print("Yayyy you nailed it")
else:
    print(f"You didn't guess correctly, the correct number is {right_number}.Try again")
```

2.  Grading Students

```python
def get_marks():
    while True:
        try:
            maths = float(input("Enter your Maths mark: "))
            science = float(input("Enter your Science mark: "))
            english = float(input("Enter your English mark: "))

            # Validating the  marks
            if (maths >= 0 and maths <= 100) and (science >= 0 and science <= 100) and (english >= 0 and english <= 100):
                return maths, science, english  
            else:
                print("Invalid input! All marks must be between 0 and 100.")
        except ValueError:
            print("Invalid input! All entries must be numeric.")

maths, science, english = get_marks()

# Calculating the average
average = (maths + science + english) / 3

# Grading the students
if average == 100:
    print("Wow!! You got a perfect score!")
elif average >= 90:
    grade = "A"
    print(f"Excellent! You got an {grade}")
elif average >= 80:
    grade = "B"
    print(f"Very good! You got a {grade}")
elif average >= 70:
    grade = "c"
    print(f"Good! You got a {grade}")
elif average >= 60:
    grade = "D"
    print(f"Needs improvement! You got a {grade}")
else:
    grade = "F"
    print(F"Fail! You got an {grade}")

# When the student scores 0
if maths == 0 and science == 0 and english == 0:
    print("Oh no! You failed everything!")
```
